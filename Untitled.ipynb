```python
df_gs = pd.read_excel('Справочник кодов ГС_02.11.2012.xls')
```

```python
df = df.rename(columns={'Строка 7-ГС': 'Код 7ГС'})
df_gs
```

```python
df = df.merge(df_gs, how='left')
```

```python
df['Сборы'] = (df['Начисленная страховая премия'] - df['Возвраты'])/1000
df
```

```python
df = df.rename(columns={'Наименование подразделения': 'subdivision_name'})
```

```python
result = df.pivot_table(
            index=('Вид страхования','Тип лица'), columns='Канал продаж', values='Сборы', aggfunc='sum'
        )
```

```python
result[['Агентские продажи','МКП в агентствах','Партнеры через агентства','Перевод партнерских продаж в офисы','Стационарные офисы']] = (
    result[['Агентские продажи','МКП в агентствах','Партнеры через агентства','Перевод партнерских продаж в офисы','Стационарные офисы']].
    round()
)
```

```python
writer = pd.ExcelWriter('output.xlsx')      # сохраним полученную таблицу в excel
result.to_excel(writer)
writer.save()
```
